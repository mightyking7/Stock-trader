```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

sns.set(style='darkgrid', palette='Dark2')

f1 = 'EM_Data_April2019.csv'
f2 = 'PDB_Data_April2019.csv'

df = pd.read_csv(f1, index_col=0)
df2 = pd.read_csv(f2)

#Clean and prepare 
dcol = ['Status','Entry_Title','Deposit_Date','Detector','Energy_Filter','Citation_Authors']

df.query('Resolution != 0')
df = df.drop(dcol, axis=1)

#Add year column 
ys = pd.DatetimeIndex(df['Map_Release']).year
df['Year'] = ys

#remove certian years
df = df[df['Year'] != 2019]

```

##D)

```python
fig, ax = plt.subplots(figsize = (12,6))
pdf2 = df2[['X-ray_total','NMR_total','Cryo_total']].copy()
pdf2.index = df2.Year
pdf2.plot(kind='line',ax=ax)
```

##E)

```python
fig, ax = plt.subplots(figsize = (12,6))
pdf2 = df2[['X-ray_total','NMR_total','Cryo_total','other_total']].copy()
pdf2.index = df2.Year
pdf2.columns = ['X-ray','NMR','Cryo','Other']
pdf2 = pdf2.sort_index(ascending=True)
pdf2.plot(kind='bar', stacked=True, ax=ax)
```
